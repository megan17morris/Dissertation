```python
import pandas as pd
import numpy as np
import re
```

```python
davis_response = pd.read_excel('University Responses/Round1.xlsx')
davis_response.head(1)
```

```python
davis_response.columns
column_names = ['participant_ID', 'question_ID','Relevant', 'Correct_Info', 'Grammatical', 'Enough_Info', 'Correct_Marked', 'One_Correct', 
                'Distinct_Option', 'Plausible', 'Usable', 'LG_Progress', 'Teacher_Blooms', 'Level_feedback','improve_feedback', 'context','context_explain', 'additional_feedback']
```

```python
davis_response.head()
davis_response=davis_response.drop('Timestamp', axis = 1)
```

```python
davis_response.insert(1,'participant_ID', [1,2,3])
```

```python
partcipant_ds = davis_response.iloc[:,:2]
davis_response = davis_response.iloc[:,2:]
```

```python
num_columns = 16
num_questions = 16

question_dfs = {}

for i in range(num_questions):
    df_name =f'question_{i+1}'
    question_dfs[df_name] = davis_response.iloc[:,0: num_columns]
    davis_response.drop(davis_response.iloc[:, 0:num_columns], inplace=True, axis=1)
```

```python
for df in question_dfs.values():
    df.insert(0,'participant_ID', [1,2,3])
```

```python
questionIds= [1,4,5,7,9,13,14,16,18,20,22,23,24,25,26,27]
```

```python
for i in range(0, len(question_dfs)):
    question_id = questionIds[i]
    df_name = "question_"+str(i+1)
    question_dfs[df_name].insert(1, 'question_ID', question_id)
```

```python
question_dfs['question_1'].head()
```

```python
for i in range(0,16):
    df_name = 'question_'+str(i+1)
    question_dfs[df_name].columns = column_names
```

```python
section_questions = pd.concat(question_dfs.values(),axis=0)
```

```python
section_questions
```

```python
relabel_dict ={
    "LG_Progress" : {'Yes, to a large extent': 2,
                     'Yes, to some extent':1,
                     'No, it does not': 0},
    "Relevant" : {'Yes, the question is relevant to the excerpt':2,
                  'The question is relevant but requires knowledge beyond what is included in the text section': 1,
                  'No, the question is not relevant': 0},
    "Correct_Info" : {'Yes, all the information in the question is correct':2,
                      'Mostly, but there are minor inaccuracies': 1,
                      'No, not all information contained in the question is correct':0},
    "Grammatical" :{'Yes, the question is grammatical and well formed':2,
                    "The question contains minor grammar or syntactic errors, but these errors don't interrupt understanding of the question":1,
                     'The question contains major grammar or syntactic errors that make the question more difficult to understand':0
                   },
    "Enough_Info" :{'Yes, it provides enough information': 2,
                   'Somewhat, but additional information would be useful for clarity':1,
                    "No, it doesn't provide enough information": 0},
    "Correct_Marked" : {'Yes, there is a correct answer and it is marked as the correct answer':2,
                         'There is a correct answer but it is not given as the correct answer':1,
                        'No, there is no correct answer': 0},
    "One_Correct" :{'Yes, the question contains only one correct answer': 2,
                     'No, there are multiple correct answers.':1,
                     'No, there are no correct answers given':0},
    "Distinct_Option" : {'Yes, they are completely unique between each other': 2,
                         'Some choices are unique, some are too similar' : 1,
                         'No, they are too similar, making them repeated choices': 0},
    
    "Plausible" : {'Yes there are no obviously wrong options and all distractors target common misconceptions': 2, 
                   'Yes however the options give away the correct answer' : 1,
                   'No, there are obviously wrong option(s) and some of the answers do not target common misconceptions': 0},
    
    "Usable" : {"Yes" : 1,
                "No" : 0},
    "Teacher_Blooms" : {"Analyzing" : 4, 
                        "Applying" : 3,
                        "Understanding" : 2,
                        "Remembering" : 1
                       }
}
    
```

```python
for col, map_dict in relabel_dict.items():
    if col in section_questions.columns:
        section_questions[col] = section_questions[col].replace(map_dict)
```

```python
section_questions
```

```python
question_bank = pd.read_excel('pkaquestionbank.xlsx')
```

```python
question_bank['question_ID'] = question_bank.index +1
```

```python
question_bank
```

```python
df_merged = pd.merge(section_questions, question_bank, on='question_ID', how='left')
```

```python
df_merged.columns
```

```python
#Take only the objective and question columns to be used for supervised model
questions_only = df_merged[['object']]
```

```python
df_merged[["Teacher_Blooms","Level_feedback", "improve_feedback","context_explain","additional_feedback", "objective", "learning_goal","scenario","question","options","correct_answer","explanation"]]
```

```python
evalCols = ['Usable', 'Plausible', 'Distinct_Option', 'One_Correct',
            'Correct_Marked', 'Enough_Info', 'Correct_Info', 'Grammatical','Relevant']
df_merged['Evaluation_Score'] = df_merged[evalCols].fillna(0).sum(axis=1)
```

```python
df_merged['Evaluation_Score']
```

```python
eval_scores = df_merged.groupby('question_ID')['Evaluation_Score'].mean().reset_index()
```

```python
eval_scores
```

```python
# Calculate percentiles
percentiles = [0, 20, 40, 60, 80, 100]
eval_scores['percentile_rank'] = pd.qcut(eval_scores['Evaluation_Score'], q=5, labels=False, duplicates='drop')

# Map percentiles to ratings
rating_mapping = {
 
    4: 5,  # Top 20th percentile
    3: 4,  # 40th to 20th percentile
    2: 3,  # 60th to 40th percentile
    1: 2,  # 80th to 60th percentile
    0: 1   # Below 80th percentile
}
eval_scores['rating'] = eval_scores['percentile_rank'].map(rating_mapping)

# Drop the temporary percentile_rank column
eval_scores.drop(columns=['percentile_rank'], inplace=True)

print(eval_scores)
```

```python
df_merged = pd.merge(df_merged, eval_scores, on='question_ID', how='left')
```
