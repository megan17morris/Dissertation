```python
def clean_column_names(columns):
    return [re.sub(r'\.\d+.*$', '', col) for col in columns]

```

```python
def transformResponse(excel, participant_ID_list, question_ID_list):
    sectionResponse = pd.read_excel(excel)
    sectionResponse = sectionResponse.drop('Timestamp', axis = 1)
    sectionResponse.insert(2,'participant_ID', participant_ID_list)
    participant_ds = sectionResponse.iloc[:, :3]
    sectionResponse = sectionResponse.iloc[:, 3:]

    num_columns = 11
    num_questions = 12
    
    question_dfs = {}
    # Create dataframes 
    for i in range(num_questions):
        df_name =f'question_{i+1}'
        question_dfs[df_name] = sectionResponse.iloc[:,0: num_columns]
        sectionResponse.drop(sectionResponse.iloc[:, 0:num_columns], inplace=True, axis=1)
    num_columns = 12
    num_questions = 4
    for i in range(num_questions):
        df_name =f'question_{i+13}'
        question_dfs[df_name] = sectionResponse.iloc[:,0: num_columns]
        sectionResponse.drop(sectionResponse.iloc[:, 0:num_columns], inplace=True, axis=1)
    # Add the questions that are left to the participant ds    
    participant_ds= pd.merge(participant_ds, sectionResponse, left_index=True, right_index=True)

    for df in question_dfs.values():
      
        df.insert(0,'participant_ID', participant_ID_list)
    # Clean the columns so they don't inclue the .num
    for df in question_dfs.values():
        df.columns = clean_column_names(df.columns)
    # For all of the questions, add their question id
    for i in range(0, len(question_dfs)):
        question_id = question_ID_list[i]
        df_name = "question_"+str(i+1)
        question_dfs[df_name].insert(1, 'question_ID', question_id)

    column_names = ['participant_ID', 'question_ID','Relevant', 'Correct_Info', 'Grammatical', 'Enough_Info', 'Correct_Marked', 'One_Correct', 
                'Distinct_Option', 'Plausible','Misconceptions', 'Usable', 'Teacher_Blooms']
    long_column_names = ['participant_ID', 'question_ID','LG_Progress', 'Relevant', 'Correct_Info', 'Grammatical', 'Enough_Info', 'Correct_Marked', 'One_Correct', 
                'Distinct_Option', 'Plausible','Misconceptions', 'Usable', 'Teacher_Blooms']
    # Adjust the column names to be shorter and consistent
    for i in range(0,12):
        df_name = 'question_'+str(i+1)
        question_dfs[df_name].columns = column_names

    for i in range(12,16):
        df_name = 'question_'+str(i+1)
        question_dfs[df_name].columns = long_column_names

    # Add LG progress to the shorter questions
    for i in range(0,12):
        df_name = 'question_'+str(i+1)
        question_dfs[df_name].insert(2, 'LG_Progress', np.NaN)

    sectionResponse = pd.concat(question_dfs.values(), axis=0)
    return participant_ds, sectionResponse
    
```

```python
# Response relabeling
relabel_dict ={
    "LG_Progress" : {'Yes, to a large extent': 2,
                     'Yes, to some extent':1,
                     'No, it does not': 0},
    "Relevant" : {'Yes, the question is relevant to the excerpt':2,
                  'The question is relevant but requires knowledge beyond what is included in the text section': 1,
                  'No, the question is not relevant': 0},
    "Correct_Info" : {'Yes, all information in the question is correct':2,
                      'Mostly, but there are minor inaccuracies': 1,
                      'No, not all information contained in the question is correct':0},
    "Grammatical" :{'Yes, the question is grammatical and well formed':2,
                    'The question contains minor grammar or syntactic errors, but these errors don’t interrupt understanding the question.':1,
                     'The question contains major grammar or syntactic errors that make the question more difficult to understand.':0
                      },
    "Enough_Info" :{'Yes it provides enough information': 2,
                   'Somewhat, but additional information would be useful for clarity':1,
                    'No, it doesn’t provide enough information': 0},
    "Correct_Marked" : {'Yes, there is a correct answer and it is marked as the correct answer':2,
                        'There is a correct answer but it is not given as the correct answer”':1,
                        'No, there is no correct answer': 0},
    "One_Correct" :{'Yes the question contains only one correct answer': 2,
                     'No, there are multiple correct answers':1,
                     'No, there are no correct answers given':0},
    "Distinct_Option" : {'Yes, they are completely unique between each other': 2,
                         'Some choices are unique, some are too similar' : 1,
                         'No, they are too similar, making them repeated choices': 0},
    
    "Plausible" : {'Yes there are no obviously wrong options': 2, 
                   'Yes however the options give away the correct answer' : 1,
                   'No, there are obviously wrong option(s)': 0},
    "Misconceptions" : {'Yes, all wrong answer choices target common misconceptions': 2,
                        'Some wrong answer choices target common misconceptions': 1,
                        'No, the wrong answer choices do not target misconceptions': 0},
    "Usable" : {"Yes" : 1,
                "No" : 0},
    "Teacher_Blooms" : {"Analyzing" : 4, 
                        "Applying" : 3,
                        "Understanding" : 2,
                        "Remembering" : 1
                       }
}
    
```

```python
relabel_dict
```

```python
for col, map_dict in relabel_dict.items():
    if col in questionEvaluation.columns:
        questionEvaluation[col] = questionEvaluation[col].replace(map_dict)
```
