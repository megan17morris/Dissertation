# Data Cleaning


This notebook will be used to convert each of the 3 datasets into datasets with a question stem and numeric label.

This notebook will also display the average number of words in the questions of each dataset and the number of questions and labels for each dataset.

```python
import pandas as pd
```

```python
# Hadifar dataset first
hadifar_train = pd.read_json('hadifar_train.json')
hadifar_val = pd.read_json('hadifar_validation.json')
```

```python
hadifar_train
```

```python
#This is how to access a list of questions
hadifar_train['questions'][0]
```

```python
# Create empty dataframe
hadifar = pd.DataFrame(columns=['question','label'])
```

```python

for chapter in hadifar_train['questions']:
    for question in chapter:
        if question['bloom'] !=  None:
            row = {'question': question['question']['normal_format'], 'label':question['bloom']}
            hadifar = pd.concat([hadifar, pd.DataFrame([row])], ignore_index=True)


for chapter in hadifar_val['questions']:
    for question in chapter:
        if question['bloom'] !=  None:
            row = {'question': question['question']['normal_format'], 'label':question['bloom']}
            hadifar = pd.concat([hadifar, pd.DataFrame([row])], ignore_index=True)



```

```python
hadifar
```

```python
#Link to dataset https://arxiv.org/pdf/2210.06104
hadifar['label'].value_counts()
```

```python
#903 Questions total
## 660 Remember
## 114 Understand
## 110 Apply
# 19 Analyse
```

```python
# Next we want all labels to be 0, 1, 2, 3
hadifar['label'] = hadifar['label'].apply(lambda x: int(x) - 1)
```

```python
hadifar['label'].value_counts()
```

```python
#Hadifar dataset has been created
```

## Yaha

```python
yaha = pd.read_csv('Yaha.csv')
```

```python
yaha = yaha.rename(columns={'QUESTION':'question',"BT LEVEL": "label"})
```

```python
yaha
```

```python
yaha['label'].unique()
```

```python
# Remove columns with evaluation and synthesis
yaha = yaha[~yaha['label'].isin(['Evaluation', 'Synthesis'])]
```

```python
yaha['label'].unique()
```

```python
label_mapping = {
    'Knowledge': 0,
    'Comprehension': 1,
    'Application': 2,
    'Analysis': 3
}
```

```python
yaha['label'] = yaha['label'].replace(label_mapping)

# Display the resulting DataFrame
print(yaha)
```

```python
yaha['label'].value_counts()
```

```python
# 400 questions, 100 for each level
# Yaha has been Created
```

## Kaggle dataset

```python
kaggle = pd.read_csv('kaggle.csv')
```

```python
kaggle.head()
```

```python
kaggle = kaggle.rename(columns={'Question':'question',"Label": "label"})
```

```python
kaggle['label'].unique()
```

```python
kaggle = kaggle[~kaggle['label'].isin(['Evaluate', 'Create'])]
```

```python
label_mapping = {
    'Remember': 0,
    'Understand': 1,
    'Apply': 2,
    'Analyse': 3
}
```

```python
kaggle['label'] = kaggle['label'].replace(label_mapping)
```

```python
kaggle['label'].value_counts()
```

```python
## 1171 rows. 271 Remember, 300 Understand, 300 Apply, 300 Analyse,
```

```python
# The Kaggle Dataset is ready
```

## Combine all Datasets

```python
combined_df = pd.concat([kaggle, yaha, hadifar], ignore_index=True)
```

```python
combined_df
```

```python
#Convert questions to lower case then strip white space
combined_df['question'] = combined_df['question'].str.lower().str.strip()

# Remove duplicate rows based on the cleaned 'question' column
combined_df = combined_df.drop_duplicates(subset='question')
```
