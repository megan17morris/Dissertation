```python
def performancePrinter(test_y, pred_y, y_prob=None):
    # Print accuracy score
    print("Accuracy Score -> ", accuracy_score(test_y, pred_y))
    
    # Print Cohen's Kappa Score
    print("Kappa Score -> ", cohen_kappa_score(test_y, pred_y))
    
    # Check if ROC AUC can be calculated (requires probability scores)
    if y_prob is not None:
        try:
            # Calculate ROC AUC score for multi-class classification
            print("ROC AUC Score -> ", roc_auc_score(test_y, y_prob, multi_class='ovr', average='macro'))
        except ValueError as e:
            print(f"Error calculating ROC AUC Score: {e}")
    else:
        print("ROC AUC Score -> Cannot calculate (probability scores required)")
    
    # Print F1 Score with macro averaging for multi-class classification
    print("F1 Score -> ", f1_score(test_y, pred_y, average='macro'))
    
    # Print classification report
    print("Classification report -> \n", classification_report(test_y, pred_y))


    # Create DataFrame from true and predicted labels
    ratings_df = pd.DataFrame({
        'True': test_y,
        'Predicted': pred_y
    })
    
    # Create a contingency table for Fleiss' Kappa
    num_classes = len(np.unique(test_y))
    contingency_table = pd.DataFrame(np.zeros((len(test_y), num_classes), dtype=int))
    
    for i in range(len(test_y)):
        true_label = test_y[i]
        pred_label = pred_y[i]
        contingency_table.loc[i, true_label] += 1
        contingency_table.loc[i, pred_label] += 1

    # Compute Fleiss' Kappa
    try:
        kappa_fleiss = fleiss_kappa(contingency_table)
        print(f"Fleiss' Kappa -> {kappa_fleiss}")
    except Exception as e:
        print(f"Error calculating Fleiss' Kappa: {e}")
```

```python
# Desired order: Remember, Understand, Apply, Analyze
ordered_labels = ["Remember", "Understand", "Apply", "Analyze"]

# Create a mapping of the labels to the correct indices
label_to_index = {label: i for i, label in enumerate(ordered_labels)}

# Fix the target offset to align with the correct indices (target indices start at 4)
target_offset = 4

# Explicit remapping to enforce the correct order on both sides
# Left side (source)
new_source = [label_to_index[labels[src]] for src in source]

# Right side (target) with the offset
new_target = [label_to_index[labels[tgt - target_offset]] + target_offset for tgt in target]

# Use the same color definitions
flow_color_map = {
    0: 'rgba(31, 119, 180, 0.3)',  # Lighter blue for Remember
    1: 'rgba(255, 127, 14, 0.3)',  # Lighter orange for Understand
    2: 'rgba(44, 160, 44, 0.3)',   # Lighter green for Apply
    3: 'rgba(214, 39, 40, 0.3)'    # Lighter red for Analyze
}

node_color_map = {
    0: 'rgb(31, 119, 180)',  # Solid blue for Remember
    1: 'rgb(255, 127, 14)',  # Solid orange for Understand
    2: 'rgb(44, 160, 44)',   # Solid green for Apply
    3: 'rgb(214, 39, 40)'    # Solid red for Analyze
}

default_color = 'rgb(128, 128, 128)'  # Gray as the fallback color

# Apply colors to links based on the new source
link_colors = [flow_color_map.get(src, default_color) for src in new_source]

# Apply colors to both left and right nodes (based on new mapping)
node_colors = [node_color_map.get(src, default_color) for src in range(len(ordered_labels))] + \
              [node_color_map.get(tgt - target_offset, default_color) for tgt in range(target_offset, target_offset + len(ordered_labels))]

# Create the Sankey diagram with enforced order
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=ordered_labels + ordered_labels,  # Duplicating the ordered labels for both source and target
        color=node_colors  # Apply colors to both sides based on the enforced order
    ),
    arrangement='snap',
    link=dict(
        source=new_source,  # Use the remapped source indices
        target=new_target,  # Use the remapped target indices
        value=value,
        color=link_colors,  # Apply the lighter colors based on the new source objective
        customdata=value,  # Pass the flow values for labeling
        hovertemplate='Value: %{value}<br>Source: %{source.label}<br>Target: %{target.label}<extra></extra>',  # Customize hover info
        label=ordered_labels  # This automatically labels the lines with the respective values
    )
))

# Update the layout to make the diagram narrower and boxes white
fig.update_layout(
    title_text="Generated Bloom's Levels to Teacher Predictions",
    font=dict(size=10, color='black'),
    title_x=0.5,  # Center the title
    title_y=0.85,  # Adjust the vertical position of the title
    width=600,  # Adjust this value to make the diagram narrower
    height=400,  # You can adjust the height as needed
    annotations=[
        dict(
            x=-0.1,
            y=-0.1,
            xref='paper',
            yref='paper',
            text='LLM-level',
            showarrow=False,
            font=dict(size=12, color='black'),
            align='center'
        ),
        dict(
            x=1.1,
            y=-0.1,
            xref='paper',
            yref='paper',
            text='Teacher-Level',
            showarrow=False,
            font=dict(size=12, color='black'),
            align='center'
        )
    ],
    paper_bgcolor='rgba(255,255,255,0)',  # Set background to white
    plot_bgcolor='rgba(255,255,255,0)'  # Set plot background to white
)

# Show the diagram
fig.show()
```
